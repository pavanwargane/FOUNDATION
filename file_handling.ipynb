###FILE HANDLING

```python
### reading whole file
with open('fie.txt','r') as file:
    content=file.read()
    print(content)
    
```

```python
## read a file line by line
with open ('fie.txt','r') as file:
    for line in file:
        print(line.strip()) 
    
```

```python
#writing a file(overwriting)

with open ('fie.txt','w')as file:
    file.write('hello  word \n')
    file.write('this is a new line')
```

```python
## write a file without overwriting
with open ('fie.txt','a') as file:
    file.write("append operation taking place \n")
```

```python
### writing a list of lines to a file
lines=['first line \n','second line \n', ' third line \n']
with open ('fie.txt','a')as file:
    file.writelines(lines)
```

###bin files

```python
## writing to a binary file 
data = b'\x00\x01\x02\x03\x04'
with open ('file.bin','wb') as file:
    file.write(data)
```

```python
## reading a binary file 

with open ('file.bin','rb') as file:
    content=file.read()
    print(content)
```

```python
### read the content from source textfile and write it to destination text file
### copying a text file 

with open ('fie.txt','r')as source_file:
    content=source_file.read()


with open ('destination.txt','w')as destination_file:
    destination_file.write(content)

    
```

```python
### writing and then reading a file 

with open ('fie.txt','w+')as file:
    file.write("hello world \n")
    file.write("this is a new line \n")
    
    ### move the cursor to the begining

    file.seek(0)
    content=file.read()
    print(content)
```
