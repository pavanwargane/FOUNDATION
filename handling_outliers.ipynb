### handling otliers

```python
# maximum,minimum,median,Q1,Q3,IQR
```

```python
import numpy as np
```

```python
lst_marks=[45,32,75,89,54,90,87,67,54,45,98,99,74,67]
minimum,Q1,median,Q3,maximum=np.quantile(lst_marks,[0.0,0.25,0.50,0.75,1.0])
```

```python
minimum,Q1,median,Q3,maximum
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_fence=Q1-1.5*(IQR)
higher_fence=Q3+1.5*(IQR)
```

```python
lower_fence
```

```python
higher_fence
```
