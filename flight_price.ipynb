### FLIGHT PRICE PREDICTION DATASET

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel('flight_price.xlsx')
df.head(5)
```

```python
df.tail()
```

```python
## basic info 
df.info()
```

```python
## feature engineering

df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df.head()
```

```python
df.info()
```

```python
## onject to int
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['dep_Hour']=df['Dep_Time'].str.split(':').str[0]
df['dep_Minutes']=df['Dep_Time'].str.split(':').str[1]
df['arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['arrival_minute']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df['dep_Hour']=df['dep_Hour'].astype(int)
df['dep_Minutes']=df['dep_Minutes'].astype(int)
df['arrival_hour']=df['arrival_hour'].astype(int)
df['arrival_minute']=df['arrival_minute'].astype(int)
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].astype(int)
```

```python
df.info()
```

```python
df=df.drop('Arrival_Time',axis=1,inplace=True)
df=df.drop('Dep_Time',axis=1,inplace=True)
df=df.drop('Route',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Du_hour']=df['Duration'].str.split('h').str[0]
```

```python
df.head(5)
```

```python
df['Du_minute']=df['Duration'].str.split('h').str[1].str.split('m').str[0]
```

```python
df.head(2)
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
df.drop('Duration',axis=1,inplace=True)
```
