```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
lbl_encoder=LabelEncoder()
```

```python
df=pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df
```

```python
encoded=lbl_encoder.fit_transform(df[['color']])
```

```python
encoded
```
